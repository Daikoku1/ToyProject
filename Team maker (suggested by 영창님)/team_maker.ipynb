### 팀 구성하기

```python
new_list = []
for i in range(len(df)):
    temp_list = []
    temp_list.append(df.iloc[i][0])
    temp_list.append(df.iloc[i][1])
    new_list.append(temp_list)
    
x = 4
grouped = []    

rep=0
while True:
    rand_idx = np.random.randint(0,len(new_list),x)
    l1 = []
    for i in rand_idx:
        l1.append(new_list[i][1])

    if len(l1) == len(set(l1)):
        print('중복없음',l1)

        team = []
        for i in sorted(rand_idx, reverse = True):
            team.append(new_list[i][0])
            del new_list[i]

        grouped.append(team)

    else:
        print('중복있음',l1)
    
    if len(new_list) < x:
        l2 = []
        for i in new_list:
            l2.append(i[1])
        if len(l2) == len(set(l2)):
            team = []
            for i in range(len(new_list)):
                team.append(new_list[i][0])
            grouped.append(team)
        elif len(l2) > len(set(l2)):
            print('fail')
            break
    
    rest = []
    for i in new_list:
        rest.append(i[1])
    
    if pd.Series(rest).value_counts().max() > len(new_list)/x or rep>10000:
        print('fail')
        print('생성된 팀 수 : ', len(grouped))
        print('남은 사람 : ', new_list)
        break
    else:
        print('팀을 생성했습니다.')
        print(grouped)
        
    if len(new_list) == 0:
        print('success')
        del grouped[-1]
        break
    rep+=1
```

# 합치기

```python
def long_func(S, n, x):
    import numpy as np
    import pandas as pd
#     S = 100
    student = list(range(1,S+1))


#     n = 20
    make_rand = np.random.randn(n)*0.02+0.1

    major_ratio = []
    for i in make_rand:
        ratio = i/sum(make_rand)
        major_ratio.append(ratio)
    major_ratio = np.array(major_ratio)*100
    major_ratio = np.around(major_ratio).astype(int)

    major_list = []
    for i in range(1,n+1):
        for j in range(major_ratio[i-1]):
            major_list.append(str(i))

    if (len(major_list) - S)>0:
        much = abs(len(major_list) - S)
        for i in range(much):
            rand_idx = np.random.randint(0,len(major_list))
            del major_list[rand_idx]
    elif (len(major_list) - S)<0:
        less = abs(len(major_list) - S)
        for i in range(less):
            rand_idx = np.random.randint(0,len(major_list))
            major_list.insert(rand_idx,major_list[rand_idx])
    else:
        pass

    df = pd.DataFrame({'student' : student,
                 'major' : major_list})

    new_list = []
    for i in range(len(df)):
        temp_list = []
        temp_list.append(df.iloc[i][0])
        temp_list.append(df.iloc[i][1])
        new_list.append(temp_list)
        
    team_num_max = S//x+1
    most_major = pd.Series(major_list).value_counts().max()

#     x = 4
    grouped = []
    rep=0
    while True:
        if most_major > team_num_max:
            print('impossible')
            break
        
        rand_idx = np.random.randint(0,len(new_list),x)
        l1 = []
        for i in rand_idx:
            l1.append(new_list[i][1])

        if len(l1) == len(set(l1)):
            print('중복없음',l1)

            team = []
            for i in sorted(rand_idx, reverse = True):
                team.append(new_list[i][0])
                del new_list[i]

            grouped.append(team)

        else:
            print('중복있음',l1)

        if len(new_list) < x:
            l2 = []
            for i in new_list:
                l2.append(i[1])
            if len(l2) == len(set(l2)):
                team = []
                for i in range(len(new_list)):
                    team.append(new_list[i][0])
                grouped.append(team)
            elif len(l2) > len(set(l2)):
                print('fail')
                break

        rest = []
        for i in new_list:
            rest.append(i[1])

        if pd.Series(rest).value_counts().max() > len(new_list)/x or rep>10000:
            print('fail')
            print('생성된 팀 수 : ', len(grouped))
            print('남은 사람 : ', new_list)
            break
        else:
            print('팀을 생성했습니다.')
            print(grouped)

        if len(new_list) == 0:
            print('success')
            del grouped[-1]
            break

        rep+=1
```
